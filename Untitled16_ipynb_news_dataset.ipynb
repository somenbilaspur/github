```python
columns = ['timestamp','brand']
combine_articles_df_interection_df = combine_articles_df_interection_df.drop(columns, axis=1)
combine_articles_ratings.head(10)
```

```python
users_interections_count_df= interection_df.groupby(['personId', 'contentId']).size().groupby('personId').size()
print('# users: %d' % len(users_interections_count_df))
users_with_enough_interections_df = users_interections_count_df[users_interections_count_df >= 5].reset_index()[['personId']]
print('# users with at least 5 interections: %d' % len(users_with_enough_interections_df))
```

```python
print('# of interections: %d' % len(interection_df))
interections_from_selected_users_df = interection_df.merge(users_with_enough_interections_df, 
               how = 'right',
               left_on = 'personId',
               right_on = 'personId')
print('# of interections from users with at least 5 interections: %d' % len(interections_from_selected_users_df))
```

```python
interections_train_df, interections_test_df= train_test_split(interection_df,
                                   stratify=interection_df['personId'],
                                   test_size=0.20,
                                   )

print('# interections on Train set: %d' % len(interections_train_df))
print('# interections on Test set: %d' % len(interections_test_df))
# interections on Train set: 31284
# interections on Test set: 7822
```

```python
#Creating a sparse pivot table with users in rows and items in columns
users_items_pivot_matrix_df = interections_train_df.pivot(index='personId', 
                                                          columns='contentId', 
                                                          values='eventStrength').fillna(0)

users_items_pivot_matrix_df.head(10)
```
